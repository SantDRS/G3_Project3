```python
# Dependencies
import requests
import json
import pandas as pd
from pprint import pprint
```

```python
# retrieve apis, and create jsons

urlRed = 'https://api.sampleapis.com/wines/reds'
urlWhite = 'https://api.sampleapis.com/wines/whites'
urlRose = 'https://api.sampleapis.com/wines/rose'
urlSparkling = 'https://api.sampleapis.com/wines/sparkling'
urlDessert = 'https://api.sampleapis.com/wines/dessert'
urlPort = 'https://api.sampleapis.com/wines/port'

redJson = requests.get(urlRed).json()
whiteJson = requests.get(urlWhite).json()
roseJson = requests.get(urlRose).json()
sparklingJson = requests.get(urlSparkling).json()
dessertJson = requests.get(urlDessert).json()
portJson = requests.get(urlPort).json()
```

```python
# convert each json into a dataFrame
reds_df = pd.DataFrame(redJson)
whites_df = pd.DataFrame(whiteJson)
rose_df = pd.DataFrame(roseJson)
sparkling_df = pd.DataFrame(sparklingJson)
dessert_df = pd.DataFrame(dessertJson)
port_df = pd.DataFrame(portJson)
```

```python
# check dataframes
reds_df.dtypes
```

```python
# add column 'wine type'
reds_df["wine type"] = 'Red'
whites_df["wine type"] = 'White'
rose_df["wine type"] = 'Rose'
sparkling_df["wine type"] = 'Sparkling'
dessert_df["wine type"] = 'Dessert'
port_df["wine type"] = 'Port'
```

```python
# check that columns were added
sparkling_df
```

```python
# create new id keys
reds_df['id'] = ['r'+ '{:0>4}'.format(str(id)) for id in reds_df['id']]
whites_df['id'] = ['w'+ '{:0>4}'.format(str(id)) for id in whites_df['id']]
rose_df['id'] = ['o'+ '{:0>4}'.format(str(id)) for id in rose_df['id']]
sparkling_df['id'] = ['s'+ '{:0>4}'.format(str(id)) for id in sparkling_df['id']]
dessert_df['id'] = ['d'+ '{:0>4}'.format(str(id)) for id in dessert_df['id']]
port_df['id'] = ['p'+ '{:0>4}'.format(str(id)) for id in port_df['id']]
```

```python
# check dataframes
whites_df
```

```python
# concatenate dataframes
wine_dataframes = [reds_df,whites_df,rose_df,sparkling_df,dessert_df,port_df]
wine_df = pd.concat(wine_dataframes)
```

```python
# check combined wine_df
wine_df
```

```python
wine_df = wine_df[['id','winery','wine','rating','location','image','average','wine type']]
```

```python
# split location into 'country' and 'region'
wine_df[['country','region']] = wine_df.location.str.split('\\n·\\n', expand=True)
```

```python
# checking dataframe
wine_df
```

```python
# split rating into 'average rating' and 'review count'
#print(type(wine_df.iloc[0]['rating']))
averageRating = []
reviewCount = []

for index, wine in wine_df.iterrows():
    avg = wine['rating']['average']
    rev = wine['rating']['reviews']
    averageRating.append(float(avg))
    reviewCount.append(int(rev[:2]))
```

```python
# adding 'average rating' and 'review count' columns
wine_df['average rating'] = averageRating
wine_df['review count'] = reviewCount
```

```python
# checking dataframe
wine_df
```

```python
# removing 'rating' and 'location' columns
del wine_df['rating']
del wine_df['location']
del wine_df['average']
```

```python
# checking dataframe
wine_df
```

```python
# creating a country_df
countryID = ['c'+'{:0>3}'.format(str(id+1)) for id in range(len(countryList))]
country_df = pd.DataFrame({'country id':countryID,'country':countryList})
country_df
```

```python
#creating a region_df
regionID = ['a'+'{:0>4}'.format(str(id+1)) for id in range(len(regionList))]
region_df = pd.DataFrame({'region id':regionID,'region':regionList})
region_df
```

```python
varietalLs = ['Pinot Noir', 'Cabernet Sauvignon', 'Merlot', 'Syrah', 'Malbec', 'Grenache', 'Chardonnay', 'Sauvignon Blanc', 'Riesling', 'Pinot Gris', 'Chenin Blanc', 'Moscato', 'Albariño', 'Viognier','Prosecco', 'Blanc de Noir', "Blanc de Blancs", 'Champagne', 'Cava','Gamay']
varietalID = wineTypeID = ['g'+str(id+1) for id in range(len(varietalLs))]
varietal_df = pd.DataFrame({'varietal id':varietalID,'varietal':varietalLs})

varietal_df
```

```python
# reasign Dfs (avoid timeout with API)
wineDf = pd.read_csv('wine.csv')
```

```python
# 
filteredDf = wineDf[wineDf['wine'].str.contains('|'.join(varietalLs))]
```

```python
# define the regular expression pattern for the varietals
pattern = rf"\b({'|'.join(varietalLs)})\b"

# use str.extract to extract the varietal from the "wine" column
filteredDf['varietal'] = filteredDf['wine'].str.extract(pattern, flags=re.IGNORECASE)
```

```python
wineryToWine = []
wineToWinery = []
for index, wine in filteredDf.iterrows():
    for jndex, winery in winery_df.iterrows():
        if wine[1] == winery[1]:
            wineryToWine.append(winery[0])
            wineToWinery.append(wine[0])
```

```python
wineWineryDict = {"wine id": wineToWinery,"winery id": wineryToWine}
wineWinery_df = pd.DataFrame(wineWineryDict)
```

```python
wineryToRegion = []
regionToWinery = []
for index, wine in filteredDf.iterrows():
    for jndex, region in region_df.iterrows():
        if wine[6] == region[1]:
            for kndex, winery in winery_df.iterrows():
                if wine[1] == winery[1]:
                    wineryToRegion.append(winery[0])
                    regionToWinery.append(region[0])
```

```python
wineryRegionDict = {"winery id": wineryToRegion, "region id": regionToWinery}
wineryRegion_df = pd.DataFrame(wineryRegionDict)
```

```python
wineryRegion_df.head()
```

```python
regionToCountry = []
countryToRegion = []
for index, wine in filteredDf.iterrows():
    for jndex, region in region_df.iterrows():
        if wine[6] == region[1]:
            for kndex, country in country_df.iterrows():
                if wine[5] == country[1]:
                    regionToCountry.append(region[0])
                    countryToRegion.append(country[0])
```

```python
regionCountryDict = {"region id": regionToCountry,"country id": countryToRegion}
regionCountry_df = pd.DataFrame(regionCountryDict)
```

```python
wineToVarietal = []
varietalToWine = []
for index, wine in filteredDf.iterrows():
    for jndex, varietal in varietal_df.iterrows():
        if wine[9] == varietal[1]:
            varietalToWine.append(varietal[0])
            wineToVarietal.append(wine[0])
```

```python
wineVarietal_df = pd.DataFrame({'wine id':wineToVarietal,'varietal id':varietalToWine})
```
